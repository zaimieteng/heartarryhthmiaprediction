```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv1D, MaxPool1D, Dense, Flatten, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import RMSprop, Adam
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPool1D, BatchNormalization, Dropout, Flatten
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

def create_cnn_model(optimizer='adam', learning_rate=1e-3, dropout=0.1):
    # Set random seed
    np.random.seed(123)
    tf.random.set_seed(123)
    
    # Input configuration
    im_shape = (X_train.shape[1], 1)
    inputs_cnn = Input(shape=im_shape, name='inputs_cnn')
    
    # Convolutional Blocks
    conv1_1 = Conv1D(64, (6), activation='relu')(inputs_cnn)
    conv1_1 = BatchNormalization()(conv1_1)
    pool1 = MaxPool1D(pool_size=(3), strides=(2), padding="same")(conv1_1)
    conv2_1 = Conv1D(64, (3), activation='relu')(pool1)
    conv2_1 = BatchNormalization()(conv2_1)
    pool2 = MaxPool1D(pool_size=(2), strides=(2), padding="same")(conv2_1)
    conv3_1 = Conv1D(64, (3), activation='relu')(pool2)
    conv3_1 = BatchNormalization()(conv3_1)
    pool3 = MaxPool1D(pool_size=(2), strides=(2), padding="same")(conv3_1)
    
    # Classification Section
    flatten = Flatten()(pool3)
    dense_end1 = Dense(64, activation='relu')(flatten)
    dense_end2 = Dense(32, activation='relu')(dense_end1)
    main_output = Dense(5, activation='softmax', name='main_output')(dense_end2)
    
    model = Model(inputs=inputs_cnn, outputs=main_output)
    
    # Configure optimizer
    if optimizer == 'rmsprop':
        opt = RMSprop(learning_rate=learning_rate)
    else:
        opt = Adam(learning_rate=learning_rate)
    
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model

# Create KerasClassifier wrapper
clf = KerasClassifier(
    build_fn=create_cnn_model,
    epochs=40,
    batch_size=32,
    verbose=1
)
```

### Final code

```python
from scikeras.wrappers import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPool1D, BatchNormalization, Dropout, Flatten
import numpy as np
import tensorflow as tf
import os
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint


def create_deterministic_cnn(optimizer='adam', learning_rate=1e-3, dropout=0.1):
    # Set environment variables for deterministic operations
    os.environ['TF_DETERMINISTIC_OPS'] = '1'
    os.environ['TF_CUDNN_DETERMINISTIC'] = '1'
    
    # Create fresh generator for reproducible random operations
    gen = tf.random.Generator.from_seed(123)
    tf.random.set_global_generator(gen)
    
    # Set seeds again inside function for complete reproducibility
    np.random.seed(123)
    tf.random.set_seed(123)
    
    # Use GlorotUniform initialization with fixed seed
    initializer = tf.keras.initializers.GlorotUniform(seed=123)
    
    # Input configuration
    im_shape = (X_train.shape[1], 1)
    inputs_cnn = tf.keras.Input(shape=im_shape, name='inputs_cnn')
    
    # First convolutional block
    x = Conv1D(64, (6), 
               kernel_initializer=initializer,
               bias_initializer='zeros',
               activation='relu')(inputs_cnn)
    x = BatchNormalization()(x)
    x = MaxPool1D(pool_size=(3), strides=(2), padding='same')(x)
    
    # Second convolutional block
    x = Conv1D(64, (3),
               kernel_initializer=initializer,
               bias_initializer='zeros',
               activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPool1D(pool_size=(2), strides=(2), padding='same')(x)
    
    # Third convolutional block
    x = Conv1D(64, (3),
               kernel_initializer=initializer,
               bias_initializer='zeros',
               activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPool1D(pool_size=(2), strides=(2), padding='same')(x)
    
    # Classification section
    x = Flatten()(x)
    
    # Dense layers with dropout for regularization
    x = Dense(64, 
             kernel_initializer=initializer,
             bias_initializer='zeros',
             activation='relu')(x)
    x = Dropout(dropout)(x)
    
    x = Dense(32,
             kernel_initializer=initializer,
             bias_initializer='zeros',
             activation='relu')(x)
    x = Dropout(dropout)(x)
    
    # Output layer
    main_output = Dense(5,
                       kernel_initializer=initializer,
                       bias_initializer='zeros',
                       activation='softmax',
                       name='main_output')(x)
    
    # Create model
    model = tf.keras.Model(inputs=inputs_cnn, outputs=main_output)
    
    # Configure learning rate schedule
    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate=learning_rate,
        decay_steps=X_train.shape[0],
        decay_rate=0.95
    )

    # Define callbacks
    callbacks = [
        EarlyStopping(
            monitor='val_loss',
            patience=8,
            restore_best_weights=True,
            min_delta=1e-4
        ),
        ModelCheckpoint(
            filepath='cnn_model.keras',
            monitor='val_loss',
            save_best_only=True,
            mode='min'
        )
    ]

    # Calculate class weights based on training distribution
    class_counts = np.sum(y_train, axis=0)
    total_samples = y_train.shape[0]
    
    # Calculate weights inversely proportional to class frequency
    class_weights = {}
    for i in range(5):
        class_weights[i] = (total_samples / (5 * class_counts[i]))
    
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr_schedule),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model, class_weights

# Create KerasClassifier wrapper
clf = KerasClassifier(
    model=create_deterministic_cnn,
    optimizer='adam',
    learning_rate=1e-3,
    dropout=0.1,
    epochs=40,
    batch_size=32,
    verbose=1
)
```

```python
from tensorflow.keras.layers import (
    Input, Conv1D, MaxPool1D, Dense, Flatten, BatchNormalization,
    Dropout
)
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import RMSprop, Adam
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.metrics import Precision, Recall
import tensorflow_addons as tfa

def create_deterministic_cnn(optimizer='adam', learning_rate=1e-3, dropout=0.1):
    # Set environment variables for deterministic operations
    os.environ['TF_DETERMINISTIC_OPS'] = '1'
    os.environ['TF_CUDNN_DETERMINISTIC'] = '1'
    
    # Create fresh generator for reproducible random operations
    gen = tf.random.Generator.from_seed(123)
    tf.random.set_global_generator(gen)
    
    # Set seeds again inside function for complete reproducibility
    np.random.seed(123)
    tf.random.set_seed(123)
    
    # Use GlorotUniform initialization with fixed seed
    initializer = tf.keras.initializers.GlorotUniform(seed=123)
    
    # Input configuration
    im_shape = (X_train.shape[1], 1)
    inputs_cnn = tf.keras.Input(shape=im_shape, name='inputs_cnn')
    
    # First convolutional block
    x = Conv1D(64, (6), 
               kernel_initializer=initializer,
               bias_initializer='zeros',
               activation='relu')(inputs_cnn)
    x = BatchNormalization()(x)
    x = MaxPool1D(pool_size=(3), strides=(2), padding='same')(x)
    
    # Second convolutional block
    x = Conv1D(64, (3),
               kernel_initializer=initializer,
               bias_initializer='zeros',
               activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPool1D(pool_size=(2), strides=(2), padding='same')(x)
    
    # Third convolutional block
    x = Conv1D(64, (3),
               kernel_initializer=initializer,
               bias_initializer='zeros',
               activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPool1D(pool_size=(2), strides=(2), padding='same')(x)
    
    # Classification section
    x = Flatten()(x)
    
    # Dense layers with dropout for regularization
    x = Dense(64, 
             kernel_initializer=initializer,
             bias_initializer='zeros',
             activation='relu')(x)
    x = Dropout(dropout)(x)
    
    x = Dense(32,
             kernel_initializer=initializer,
             bias_initializer='zeros',
             activation='relu')(x)
    x = Dropout(dropout)(x)
    
    # Output layer
    main_output = Dense(5,
                       kernel_initializer=initializer,
                       bias_initializer='zeros',
                       activation='softmax',
                       name='main_output')(x)
    
    # Create model
    model = tf.keras.Model(inputs=inputs_cnn, outputs=main_output)
    
    # Configure learning rate schedule
    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate=learning_rate,
        decay_steps=X_train.shape[0],
        decay_rate=0.95
    )

    # Define callbacks
    callbacks = [
        EarlyStopping(
            monitor='val_loss',
            patience=8,
            restore_best_weights=True,
            min_delta=1e-4
        ),
        ModelCheckpoint(
            filepath='cnn_model.keras',
            monitor='val_loss',
            save_best_only=True,
            mode='min'
        )
    ]

    # Calculate class weights based on training distribution
    class_counts = np.sum(y_train, axis=0)
    total_samples = y_train.shape[0]
    
    # Calculate weights inversely proportional to class frequency
    class_weights = {}
    for i in range(5):
        class_weights[i] = (total_samples / (5 * class_counts[i]))
    
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr_schedule),
        loss='categorical_crossentropy',
        metrics=['accuracy',
                tfa.metrics.F1Score(
                num_classes=5,
                average='macro',
                threshold=0.5
            )
        ]
    )   
    
    return model, class_weights

# Create KerasClassifier wrapper
clf = KerasClassifier(
    model=create_deterministic_cnn,
    optimizer='adam',
    learning_rate=1e-3,
    dropout=0.1,
    epochs=40,
    batch_size=32,
    verbose=1
)
```
